```python
#공통 경로
path = '/content/drive/MyDrive/usg_2023/usg1_dataset/usg1_dataset/'
```

```python
submit = pd.read_csv(path + 'sample_submission.csv')
submit.head()
```

**모델 로드**

```python
#best모델은 앞서 저장한 모델이 아닌 모델/BEST폴더 안에 따로 저장
model = tf.keras.models.load_model(path + "문제1/모델/BEST/best_model.h5")
```

**결과 및 결언**

```python
for i  in tqdm(range(submit.shape[0])):
    filename = submit.loc[i,'image_name']
    img = cv2.imread(path +'test_images/' + filename)
    resized = cv2.resize(img, (150, 150))
    test = resized.reshape(1,150,150,3)
    y_pred = model.predict(test)
    y_pred_num = np.argmax(y_pred, axis=1)[0]
    submit.loc[i,'label'] = y_pred_num
```
